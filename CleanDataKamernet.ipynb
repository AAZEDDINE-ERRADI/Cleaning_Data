```python
# View Table Data
```

```python
import pandas as pd
df = pd.read_excel('Kamerent.xlsx', index_col =0)
df
```

```python
# Cleaning Column (Room Type)
```

```python
df['Room Type'] = df['Room Type'].str.replace("-","")
df
```

```python
# Cleaning Column (Rent Price)
```

```python
df['Rent Price'] = df['Rent Price'].str.replace(".","")
df
```

```python
# Cleaning Column (Surface Size)
```

```python
df['Surface Size'] = df['Surface Size'].str.replace("m2","(m2)")
df
```

```python
# # Cleaning Column (Available in date)
```

```python
df['Available in date'] = df['Available in date'].str.replace("Log in","Not Available")
df
```

```python
# Cleaning Type NAN to Not Available 
```

```python
df.fillna('Not Available')
```

```python
# Save Data as file format csv
```

```python
df.to_csv('CleanDataKamerent.csv')
```
